1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter. 

```python
import math
import numpy as np

class Counter:
    curr_val = np.nan
    
    def __init__(self, max, min = 0, inc = 1): 
        if (max < min):
            raise ValueError("max cannot be less than min")
        
        self.max = max
        self.min = min
        self.inc = inc

        self.count()
        
    def count(self):
        self.curr_val = self.min 
        while (self.curr_val < self.max):
            print(self.curr_val, end = " ")
            self.curr_val += self.inc
        print("")

    def reset(self):
        self.curr_val = np.nan
        
```

```python
import math
import numpy as np

class Counter:

    # there should be no actual value upon initial declaration
    __curr_val = np.nan
    
    def __init__(self, __max, __min = 0, __inc = 1): 
        # ensure that you can actually hit a bounds
        if (__max < __min and __inc > 0):
            raise ValueError("Your target value cannot be less than your lowerbound if you are incrementing.")
        elif (__max > __min and __inc < 0):
            raise ValueError("Your target value cannot be larger than your lowerbound if you are decrementing.")
        elif (__inc == 0):
            raise ValueError("You must be able to increment or decrement.")

        self.__max = __max
        self.__min = __min
        self.__inc = __inc

        self.count()
    
    def count (self):
        # if you wanna go up
        if (self.__inc > 0):
            self.__curr_val = self.__min

            while (self.__curr_val < self.__max):
                print(self.__curr_val, end = " ")
                self.__curr_val += self.__inc
            print()
            
        else:
            # you go DOWN
            self.__curr_val = self.__min

            while (self.__curr_val > self.__max):
                print(self.__curr_val, end = " ")
                self.__curr_val += self.__inc
            print()
    
    def reset_counter (self):
        self.__curr_val = np.nan

    def get_max (self):
        return self.__max
    
    def get_min (self):
        return self.__min
    
    def get_inc (self):
        return self.__inc
    
    def get_curr_val (self):
        return self.__curr_val
```

3. Implement a class to represent a rectangle, holding the length, width, and $x$ and $y$ coordinates of a corner of the object. Implement functions that compute the area and perimeter of the rectangle. Make all data members private and privide accessors to retrieve values of data members. 

```python
class Rectangle:

    def __init__(self, l, w, corner_coords):
        self.__L = l
        self.__W = w
        self.__coords = corner_coords

    def __repr__(self, obj):
        return f"{repr(self.obj)}"
    
    def get_area (self):
        return self.__L * self.__W

    def get_perimeter (self):
        return (self.__L * 2) + (self.__W * 2)
    
    def get_length (self):
        return self.__L

    def get_width (self):
        return self.__W
    
    def get_coordinates (self):
        return self.__coords
    
    def get_hypotenuse (self):
        return math.sqrt((self.__L ** 2) + (self.__W ** 2))
```

```python
r = Rectangle(3, 5, (5, 0))

print(f"Area: {r.get_area()}")
print(f"Coordinates: {r.get_coordinates()}")
print(f"Perimeter: {r.get_perimeter()}")
print(f"Hypotenuse: {r.get_hypotenuse()}")
```

4. Implement a class to represent a circle, holding the radius and $x$ and $y$ coordinates of center of the object. Implement functions that compute the area and perimeter of the rectangle. Make all data members private and privide accessors to retrieve values of data members. 

```python
class Circle:

    def __init__(self, r, coords):
        self.__r = r
        self.__coords = coords

    def __repr__(self):
        return f"{repr(self.obj)}"
        
    def get_area (self):
        return np.pi * (self.__r ** 2)

    def get_perimeter (self):
        return 2*np.pi*self.__r

    def get_coordinates (self):
        return self.__coords
```

6. Implement a triangle class analogous to the rectangle and circle in question 5.

```python
class triangle(shapes):

    def __init__(self, adj, opp, coords):
        self.__adj = adj
        self.__opp = opp
        self.__hyp = math.sqrt((adj ** 2) + (opp ** 2))
        self.__coords = coords

    def get_area (self):
        s = (self.__adj + self.__opp + self.__hyp)/2
        return math.sqrt(s*(s-self.__adj)*(s-self.__opp)*(s-self.__hyp))

    def get_height (self):
        return 2*(self.get_area()/self.__hyp)

    def get_perimeter (self):
        return self.__adj + self.__hyp + self.__opp

    def get_coords (self):
        return self.__coords
```

```python
t = triangle(3, 5, (7,4))

print(t.get_area())
print(t.get_height())
print(t.get_perimeter())
print(t.get_coords())
```
